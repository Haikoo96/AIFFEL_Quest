```python
# 이미지 불러오고 리사이즈 부터 정규화하는 함수
def img_normalize(path, height, width):
    image_files = os.listdir(path) # 파일 경로에 있는 폴더안에 모든 파일을 리스트 형태로 저장
    images = []
    # 폴더안에 있는 모든 파일을 하나씩 for 문으로 보기
    for file in image_files:
        img_path = os.path.join(path, file)  # 파일의 전체 경로를 구함
        img = Image.open(img_path)
        rgb_img = img.convert('RGB') # RGB 로 변환
        images.append(rgb_img) # images 리스트에 RGB 로 변환된 이미지를 추가

    # 리사이징
    resized_images = [img.resize((width, height)) for img in images]

    # 이미지 정규화 
    normalized_images = [np.array(img) / 255.0 for img in resized_images]
    return normalized_images

# 파일 경로
img_path_tulip = r"C:\Users\박해극\Downloads\flower_trainb\flowers\tulip"
img_path_sunflower = r"C:\Users\박해극\Downloads\flower_trainb\flowers\sunflower"
img_path_rose = r"C:\Users\박해극\Downloads\flower_trainb\flowers\rose"
img_path_dandelion = r"C:\Users\박해극\Downloads\flower_trainb\flowers\dandelion"
img_path_daisy = r"C:\Users\박해극\Downloads\flower_trainb\flowers\daisy"

# 파라미터 가로 세로 설정
h, w = (180, 180)
img_normalized_tulip = img_normalize(img_path_tulip, h, w)
img_normalized_sunflower = img_normalize(img_path_sunflower, h, w)
img_normalized_rose = img_normalize(img_path_rose, h, w)
img_normalized_dandelion = img_normalize(img_path_dandelion, h, w)
img_normalized_daisy = img_normalize(img_path_daisy, h, w)

img_arr_test = np.array(img_normalized_rose)

print(f"shape of test images: {img_arr_test.shape}")

# 이미지 그려보기
plt.imshow(img_normalized_rose[1])
plt.show
```

```python
# Example flower type labels (you can modify these as needed)
flower_labels = {'roses': 0, 'daisy': 1, 'dandelion': 2, 'sunflowers': 3, 'tulips': 4}

# Initialize empty lists for the combined dataset and labels
combined_testset = []
combined_test_label = []

# Function to add images and labels to the combined lists
def add_to_dataset(flower_type, images, label):
    combined_testset.extend(images)
    combined_test_label.extend([label] * len(images))

# Add each flower type to the dataset
add_to_dataset('roses', img_normalized_rose, flower_labels['roses'])
add_to_dataset('daisy', img_normalized_daisy, flower_labels['daisy'])
add_to_dataset('dandelion', img_normalized_dandelion, flower_labels['dandelion'])
add_to_dataset('sunflower', img_normalized_sunflower, flower_labels['sunflowers'])
add_to_dataset('tulips', img_normalized_tulip, flower_labels['tulips'])

x_test = np.array(combined_testset)
y_test = np.array(combined_test_label)

x_test_shape = x_test.shape
y_test_shape = y_test.shape

print(f'shape of x_test: {x_test_shape}')
print(f'shape of y_test: {y_test_shape}')
```

```python
from sklearn.model_selection import train_test_split

# Assuming you have your features and labels
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
```

```python
new_model = tf.keras.models.load_model(r"C:\Users\박해극\Aiffel\AIFFEL_Quest\DL_lms\exploration1")
```

```python
new_model.compile(optimizer='adam', 
                  loss='sparse_categorical_crossentropy', 
                  metrics=['accuracy'])

new_model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
# Define the early stopping callback
early_stopper = EarlyStopping(
    monitor='val_loss',     # Metric to be monitored 
    patience=4,            # Number of epochs with no improvement after which training will be stopped
    verbose=1,
    mode='auto',            # Decides whether the monitored quantity is supposed to increase or decrease
    restore_best_weights=True  # Whether to restore model weights from the epoch with the best value of the monitored quantity
)

# Assume 'model' is a previously defined Keras model
# Assume 'train_data' and 'val_data' are the training and validation data
history = new_model.fit(
    x_train,
    y_train,
    validation_data=(x_val, y_val),
    epochs=10,
    callbacks=[early_stopper]  # Include the early stopping callback
)
```

```python
history_dict = history.history

loss = history_dict['loss']
val_loss = history_dict['val_loss']

epochs = range(1, len(loss)+1)
fig = plt.figure(figsize = (12,5))

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(epochs, loss, 'b-', label='train_loss')
ax1.plot(epochs, val_loss, 'r-', label='val_loss')
ax1.set_title('Train and Validation Loss')
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Loss')
ax1.grid()
ax1.legend()

accuracy = history_dict['accuracy']
val_accuracy = history_dict['val_accuracy']

ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(epochs, accuracy, 'b-', label='train_accuracy')
ax2.plot(epochs, val_accuracy, 'r-', label='val_accuracy')
ax2.set_title('Train and Validation Accuracy')
ax2.set_xlabel('Epochs')
ax2.set_ylabel('Accuracy')
ax2.grid()
ax2.legend()

plt.show()
```

```python
new_model.evaluate(x_test, y_test)
```
